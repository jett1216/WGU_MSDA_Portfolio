```python
#Calculate Z-score for 'Population'

df['Z_Score_Population']=stats.zscore(df['Population'])
df[['Population','Z_Score_Population']].head(40)
```

```python
#Print number of outliers based on z-score for 'Population'

df_outliers = df.query('(Z_Score_Population > 3) | (Z_Score_Population < -3)')
print("Number of outliers:", df_outliers.shape[0])
print("Outlier values:", df_outliers['Population'].to_dict().values())
print("min=",min(df_outliers['Population']))
print("max=",max(df_outliers['Population']))
```

```python
# Set the style of seaborn
sns.set(style='whitegrid')

#Plot for 'Population' Distribution
sns.histplot(df['Population'])

#Annotate skewness
from scipy.stats import skew 
print(skew(df['Population'], axis=0,))

plt.show()
```

```python
#Applying Yeo-Johnson transformation
from scipy.stats import yeojohnson
df['YeoJohnson_Population'], _ = yeojohnson(df['Population']) 

#Print skewness after Yeo-Johnson transformation
print(f"Skewness after Yeo-Johnson Transformation: {df['YeoJohnson_Population'].skew():.5f}")
```

```python
# Calculate skewness
Population_skew = df['Population'].skew()
YeoJohnson_Population_skew = df['YeoJohnson_Population'].skew()
 
# Set the style of seaborn
sns.set(style='whitegrid')
 
# Create a figure for 2 subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
 
# Plot for Population (positively skewed)
sns.histplot(df['Population'], kde=True, ax=ax[0], color='skyblue')
ax[0].set_title('Original Distribution of Population (Positive/Right Skew)', fontsize=16)
ax[0].set_xlabel('Population')
ax[0].set_ylabel('Count')

# Annotate Skewness
ax[0].text(0.5, 0.5, f'Skew: {Population_skew:.2f}', transform=ax[0].transAxes,
           horizontalalignment='right', color='black', weight='bold',
           fontsize=14)
 
# Plot for Yeo-Johnson Population
sns.histplot(df['YeoJohnson_Population'], kde=True, ax=ax[1], color='salmon')
ax[1].set_title('Yeo-Johnson Population Distribution', fontsize=16)
ax[1].set_xlabel('Yeo-Johnson Population')
ax[1].set_ylabel('Count')

# Annotate Skewness
ax[1].text(0.5, 0.5, f'Skew: {YeoJohnson_Population_skew:.2f}', transform=ax[1].transAxes,
           horizontalalignment='right', color='black', weight='bold',
           fontsize=14)
 
plt.tight_layout()
plt.show()
```

```python
#Prepare for PCA by normalizing data

from sklearn.decomposition import PCA

test_pca = df[['Population', 'Income', 'Outage_sec_perweek', 'Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year']]

test_pca_normalized=(test_pca-test_pca.mean())/test_pca.std()

pca = PCA(n_components=test_pca.shape[1])

pca.fit(test_pca_normalized)

```

```python
#Perform PCA

PCA(n_components=6)

test_pca2=pd.DataFrame(pca.transform(test_pca_normalized),columns=['PC1','PC2','PC3', 'PC4', 'PC5', 'PC6'])

loadings = pd.DataFrame(pca.components_.T, columns=['PC1','PC2','PC3','PC4', 'PC5', 'PC6'], 
index=test_pca_normalized.columns)

test_pca2


```

```python
#Selecting PCs utilizing Kaiser Rule (Kaiser Criterion)

#Calculate covariance matrix
cov_matrix=np.dot(test_pca_normalized.T, test_pca_normalized) / test_pca.shape[0]



```

```python
#Create eigenvalues

eigenvalues=[np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector))for eigenvector in pca.components_]
```
